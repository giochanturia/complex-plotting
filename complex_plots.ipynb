```python
import numpy as np
import matplotlib.pyplot as plt
import ComplexPlotting as CP
```

```python
def square_root(z):
    return np.sqrt(z + 0j)


def square_root_on_the_second_sheet(z):
    return -np.sqrt(z + 0j)


sqrt_z_I = square_root(CP.z_CP)
sqrt_z_II = square_root_on_the_second_sheet(CP.z_CP)

curve_sqrt_z_cut_I = CP.extra_curve(CP.z_re, np.zeros(CP.z_re.shape), square_root(CP.z_re + 1j * CP.y_near_zero))
curve_sqrt_z_cut_II = CP.extra_curve(CP.z_re, np.zeros(CP.z_re.shape), square_root_on_the_second_sheet(CP.z_re - 1j * CP.y_near_zero))
```

```python
plot_sqrt_z_I = CP.plot_complex(
    CP.z_RE, CP.z_IM, sqrt_z_I, label=r"(\sqrt{z})^{I}", extra_curves=[curve_sqrt_z_cut_I]
)
```

```python
plot_sqrt_z_II = CP.plot_complex(
    CP.z_RE, CP.z_IM, sqrt_z_II, label=r"(\sqrt{z})^{II}", extra_curves=[curve_sqrt_z_cut_II]
)
```

```python
def logarithm(z):
    return np.log(z + 0j)


def logarithm_on_the_second_sheet(z):
    return logarithm(z) + 2 * np.pi * 1j


log_z_I = logarithm(CP.z_CP)
log_z_II = logarithm_on_the_second_sheet(CP.z_CP)

curve_log_z_cut_I = CP.extra_curve(CP.z_re, np.zeros(CP.z_re.shape), logarithm(CP.z_re + 1j * CP.y_near_zero))
curve_log_z_cut_II = CP.extra_curve(CP.z_re, np.zeros(CP.z_re.shape), logarithm_on_the_second_sheet(CP.z_re - 1j * CP.y_near_zero))
```

```python
plot_log_z_I = CP.plot_complex(
    CP.z_RE, CP.z_IM, log_z_I, label=r"(\log{z})^{I}", extra_curves=[curve_log_z_cut_I]
)
```

```python
plot_log_z_II = CP.plot_complex(
    CP.z_RE, CP.z_IM, log_z_II, label=r"(\log{z})^{II}", extra_curves=[curve_log_z_cut_II]
)
```

```python
plot_sqrt_z_I.savefig("img/sqrt_z_I.png", dpi=300)
plot_sqrt_z_II.savefig("img/sqrt_z_II.png", dpi=300)
plot_log_z_I.savefig("img/log_z_I.png", dpi=300)
plot_log_z_II.savefig("img/log_z_II.png", dpi=300)
```
